```python
df.head()
```

```python
df_n=df[['Discount Amount (INR)','Gross Amount','Net Amount']]
```

```python
corr = df_n.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import plotly.graph_objs as go

scatter = go.Scatter3d(
    x=df['Discount Amount (INR)'],
    y=df['Gross Amount'],
    z=df['Net Amount'],
    mode='markers',
    marker=dict(
        size=5,
        color='blue',
        opacity=0.8
    )
)

layout = go.Layout(
    title='3D Scatter Plot',
    scene=dict( # moved scene parameter to layout
        xaxis_title='Discount Amount (INR)',
        yaxis_title='Gross Amount',
        zaxis_title='Net Amount'
    )
)

fig = go.Figure(data=[scatter], layout=layout)
fig.show()
```
